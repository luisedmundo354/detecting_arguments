# Dataset Statistics Report

This notebook computes descriptive statistics for the adjudicated gold set and the clean annotation set, prints compact summaries, saves full text reports, and writes sentence-level JSONL exports.

```python
import sys
from pathlib import Path
import json

from dataset_statistics import (
    build_sentence_dataset,
    compute_dataset_statistics,
    describe_annotation_schema,
    save_text_report,
)

ROOT = Path.cwd()
GOLD_DIR = ROOT / "annotations" / "final_annotations_gold"
CLEAN_DIR = ROOT / "annotations" / "final_annotations_gregory_zaid_clean"
OUTPUT_DIR = ROOT / "dataset_statistics"

print(f"Repo root: {ROOT}")
print(f"Gold dir: {GOLD_DIR}")
print(f"Clean dir: {CLEAN_DIR}")
print(f"Output dir: {OUTPUT_DIR}")
```

```python
schema = describe_annotation_schema([GOLD_DIR, CLEAN_DIR])
print(json.dumps(schema, indent=2))
```

```python
gold_stats = compute_dataset_statistics(
    GOLD_DIR,
    dataset_name="final_annotations_gold",
    view="export",
)
gold_report_path = save_text_report(gold_stats)

print("GOLD SUMMARY")
print(json.dumps(gold_stats.summary_metrics, indent=2))
print("Gold label distribution:")
print(json.dumps(gold_stats.label_distribution, indent=2))
print(f"Saved gold report to: {gold_report_path}")
```

```python
clean_export_stats = compute_dataset_statistics(
    CLEAN_DIR,
    dataset_name="final_annotations_gregory_zaid_clean",
    view="export",
)
clean_export_report_path = save_text_report(clean_export_stats)

print("CLEAN EXPORT SUMMARY")
print(json.dumps(clean_export_stats.summary_metrics, indent=2))
print("Clean export label distribution:")
print(json.dumps(clean_export_stats.label_distribution, indent=2))
print(f"Saved clean export report to: {clean_export_report_path}")

clean_case_stats = compute_dataset_statistics(
    CLEAN_DIR,
    dataset_name="final_annotations_gregory_zaid_clean",
    view="case",
)
clean_case_report_path = save_text_report(clean_case_stats)

print("CLEAN CASE-COLLAPSED SUMMARY")
print(json.dumps(clean_case_stats.summary_metrics, indent=2))
print("Clean case-collapsed label distribution:")
print(json.dumps(clean_case_stats.label_distribution, indent=2))
print(f"Saved clean case-collapsed report to: {clean_case_report_path}")
```

```python
gold_sentences_path = OUTPUT_DIR / "final_annotations_gold_export_sentences.jsonl"
clean_export_sentences_path = OUTPUT_DIR / "final_annotations_gregory_zaid_clean_export_sentences.jsonl"
clean_case_sentences_path = OUTPUT_DIR / "final_annotations_gregory_zaid_clean_case_sentences.jsonl"

gold_sentences = build_sentence_dataset(
    GOLD_DIR,
    dataset_name="final_annotations_gold",
    view="export",
    save_jsonl=True,
    output_path=gold_sentences_path,
)
clean_export_sentences = build_sentence_dataset(
    CLEAN_DIR,
    dataset_name="final_annotations_gregory_zaid_clean",
    view="export",
    save_jsonl=True,
    output_path=clean_export_sentences_path,
)
clean_case_sentences = build_sentence_dataset(
    CLEAN_DIR,
    dataset_name="final_annotations_gregory_zaid_clean",
    view="case",
    save_jsonl=True,
    output_path=clean_case_sentences_path,
)

print(f"Gold sentence records: {len(gold_sentences)} -> {gold_sentences_path}")
print(f"Clean export sentence records: {len(clean_export_sentences)} -> {clean_export_sentences_path}")
print(f"Clean case sentence records: {len(clean_case_sentences)} -> {clean_case_sentences_path}")
print("Sample clean case sentence record:")
print(json.dumps(clean_case_sentences[0].to_dict(), indent=2))
```
