```python
import json
from pathlib import Path
from labelstudio import render_annotation_tree, DEFAULT_OUTPUT_DIR

ROOT = Path("..").resolve()
ANNOTATIONS_DIR = ROOT / "annotations"
OUTPUT_DIR = ROOT / DEFAULT_OUTPUT_DIR  #defaults to annotations_tree_visualization
OUTPUT_DIR.mkdir(exist_ok=True)

for annotation_path in sorted(ANNOTATIONS_DIR.glob("*.json")):
    with annotation_path.open(encoding="utf-8") as fp:
        annotation = json.load(fp)

    output_path = render_annotation_tree(
        annotation,
        output_dir=str(OUTPUT_DIR),
        fmt="pdf",         # swap to"png"/"svg" if you prefer
        invert_relations=True,
        rankdir="TB",      # set to "LR" for left-to-right layouts
    )
    print(f"Rendered {annotation_path.name} -> {output_path}")
```
